### The Type of Data

There are basically two types of data: 

1. Categorical Data or Qualitative Data 
2. Numerical Data or Quantitative Data

The Categorical Data is a kind of data which is a categorical measurement expressed not in terms of number. It represents characteristics such as name, location, gender,etc. Also, ti can tak on numerical values (for example, '1' represents male and '0' represents female) It can be seperated into "Nominal Data" and "Ordinal Data".

* Nominal values provide only enough information to distinguish on object from another;
* Ordinal values provide information to order objects

Numerical Data is a numerical measurement expressed not by means of a natural language description, instead, in terms of numbers. It can be seperated into "Interval Data" and "Ratio"

* Interval Data: the differeneces between values are meaningful;
* Ratio: both differences and ratios are meaningful

| Attribute Type | Example |
| :---: | :----: |
| Nominal | zip codes, Student ID number |
| Ordinal | Street numbers, students' grades |
| Interval | calendar dataes, temperature (Celsius or Fahrenheit )|
| Ratio | electrical current, temperature (Kelvin) |

There are multiple different types of data: Graph-Based Data, Time Series Data, Sequential Data, etc.

##### Outliers 

**Outliers** are data objects with characteristics that are considerably different than most of the other data objects in the data set

* Case 1: Outliers are nosie that interferes with data analysis 
* Case 2: Outliers are the goal of our analysis 
    * Credit card fraud
    * Intrusion dection
    

```r
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)
data_set <- as.data.frame(x = x, y = y)
relation <- lm(y~x)
plot(x,y,col = "blue",main = "Height & Weight Regression(Without Outlier)")
abline(relation)
```

The above plot shows the regression model of Height & Weight Regression, if we add one outliers into the dataset, the whole model would change:

```r
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131,10)
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48, 100)
relation <- lm(y~x)
par(mfrow=c(3, 1))
plot(data_set$x, data_set$y, main = "data without outlier")
plot(x, y, main = "data with outlier")
plot(x,y,col = "blue",main = "Height & Weight Regression(with Outlier)")
abline(relation)
```

How to handle with missing values? In fact, there are multiple ways. The easiest way is to remove all rows or columns that containing a missing values; or we can use some technique to estimate them.

* Handling missing values:
    * Eliminate data objects or variables
    * Estimate missing values

##### Similarity and Dissimilarity Measures

* Similarity Measure
    * Numerical measure of how alike two data objects are
    * Is higher when objects are more alike
    * Often falls in the range $\left[0, 1\right]$
* Dissimilarity measure
    * Numerical measure of how different two data object are
        * Lower when objects are more alike
        * Minimum dissimilarity is often $0$
        * Upper limit varies
* Proximity refers to a similarity or dissimilarity

#### Distance 

How to define "close", "far", "similar" or "dissimilar"? Mathematically, we use a numerical index called "Distance". 

##### Euclidean Distance

$$d(x,y) = \sqrt{\sum_{k = 1}^n (x_k - y_k)^2}$$

where $n$ is the number of dimensions (attributes) and $x_k$ and $y_k$ are, repecitvely, the $k^{\text{th}}$ attributes (components) or data object $x$ and $y$.

Before using the Euclidean Distance, we need standarize the data if their scales are diferent.

### Association Rules & Frequent Itemsets

Associationmining is commonly used to make product recommendation and shelf planning. By identifying products that are frequently bought together, merchant may sell the product in pairs or put the frequently combination together, also, they could use these rules to manage their inventory.

#### APRIORI Algorithm 

a level-wise, breadth-first which counts trasaction to find frequent itemsets.

#### EClat 
`Eclat()` takes in a transaction object and gives the most frequent items in the databased the support you provide to the `supp` argument. The `maxlen` defines the maximum number of items in each itemset of frequent items. 

```r
install.packages("arules", dep = T, repos = 'https://mirrors.tuna.tsinghua.edu.cn/CRAN')
library(arules)
```

```r
data(Groceries)
## Mine association rules.
rules <- apriori(Groceries, 
	parameter = list(supp = 0.5, conf = 0.9, target = "rules"))
summary(rules)
```

```r
data(Groceries)
```

```r
frequentItems <- eclat (Groceries, parameter = list(supp = 0.07, maxlen = 15)) # calculates support for frequent items
inspect(frequentItems)
#>    items                         support   
#> 1  {other vegetables,whole milk} 0.07483477
#> 2  {whole milk}                  0.25551601
#> 3  {other vegetables}            0.19349263
#> 4  {rolls/buns}                  0.18393493
#> 5  {yogurt}                      0.13950178
#> 6  {soda}                        0.17437722
itemFrequencyPlot(Groceries, topN=10, type="absolute", main="Item Frequency") # plot frequent items
```

### Classification

#### Decision tree 

It is a type of supervised learning algorithm that can be used in both regression and classification problems. 
* Root Node: represents the entire population or sample. It futher gets divied into two or more homogeneous sets. 
* Splitting: is a process of dividing a node into two or more sub-nodes

```r
library(rpart)
fit <- rpart(Kyphosis ~ Age + Number + Start, data = kyphosis)
fit2 <- rpart(Kyphosis ~ Age + Number + Start, data = kyphosis,
              parms = list(prior = c(.65,.35), split = "information"))
fit3 <- rpart(Kyphosis ~ Age + Number + Start, data = kyphosis,
              control = rpart.control(cp = 0.05))
par(mfrow = c(1,2), xpd = NA) # otherwise on some devices the text is clipped
plot(fit)
text(fit, use.n = TRUE)
plot(fit2)
text(fit2, use.n = TRUE)
```

```r
library(randomForest)
## Classification:
##data(iris)
set.seed(71)
iris.rf <- randomForest(Species ~ ., data=iris, importance=TRUE,
                        proximity=TRUE)
print(iris.rf)
## Look at variable importance:
round(importance(iris.rf), 2)
## Do MDS on 1 - proximity:
iris.mds <- cmdscale(1 - iris.rf$proximity, eig=TRUE)
op <- par(pty="s")
pairs(cbind(iris[,1:4], iris.mds$points), cex=0.6, gap=0,
      col=c("red", "green", "blue")[as.numeric(iris$Species)],
      main="Iris Data: Predictors and MDS of Proximity Based on RandomForest")
par(op)
print(iris.mds$GOF)
```
